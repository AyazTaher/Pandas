<div dir="rtl" style="text-align: right;">
3 التعامل مع القيم الفارغة Missing Values Handling
 </div>

 

<div dir="rtl" style="text-align: right;">
3.1 الكشف عن القيم الفارغة
 </div>

 

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

<div dir="rtl" style="text-align: right;">
3.2 ملء القيم الفارغة بقيمة ثابتة للأعمدة الرقمية

 </div>

```python
df["Total_Visits_Last_Year"] = df["Total_Visits_Last_Year"].fillna(0)
df["Total_Visits_Last_Year"].isna().sum()
```

```python
df.isna().sum()[df.isna().sum() > 0].sort_values(ascending=False)
```

<div dir="rtl" style="text-align: right;">
3.3 ملء القيم الفارغة بقيمة ثابتة للأعمدة النصية


 </div>

```python
df["Return_Reason"] = df["Return_Reason"].fillna("not known")
df["Return_Reason"].isna().sum()
```

<div dir="rtl" style="text-align: right;">
3.4 ملء القيم الفارغة بالوسط الحسابي (Mean)

 </div>

```python
df["Quantity"].mean()
```

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].mean())
df["Quantity"].isna().sum()
```

```python
df.isna().sum()[df.isna().sum() > 0].sort_values(ascending=False)
```

<div dir="rtl" style="text-align: right;">

3.5 ملء القيم الفارغة بالمنوال (Mode)
 </div>

```python
df["Age"] = df["Age"].fillna(df["Age"].mode()[0])
df["Age"].isna().sum()
```

<div dir="rtl" style="text-align: right;">
3.6 ملء القيم الفارغة بالانحراف المعياري (Standard Deviation)

 </div>

```python
df["Credit_Score"] = df["Credit_Score"].fillna(df["Credit_Score"].std())
df["Credit_Score"].isna().sum() 
```

<div dir="rtl" style="text-align: right;">
3.7 ملء القيم الفارغة باستخدام شرط (Conditional Filling)

 </div>

```python
df["Monthly_Spending_Estimate"] = df["Monthly_Spending_Estimate"].fillna(df["Monthly_Spending_Estimate"].apply(lambda x: 3500 if x < 2500 else 400))
df["Monthly_Spending_Estimate"].isna().sum()
```

```python
df.isna().sum()[df.isna().sum() > 0].sort_values(ascending=False)
```

<div dir="rtl" style="text-align: right;">


 </div>

<div dir="rtl" style="text-align: right;">
3.8 تطبيق bfill على أعمدة محددة فقط

 </div>

```python
df['Customer_Tier'] =  df['Customer_Tier'].ffill()
df['Customer_Tier'].isna().sum()
```

<div dir="rtl" style="text-align: right;">


 </div>

```python
df['Preferred_Contact_Time'] = df['Preferred_Contact_Time'].bfill()
df['Preferred_Contact_Time'].isna().sum()
```

<div dir="rtl" style="text-align: right;">


 </div>

```python
df.isna().sum()[df.isna().sum() > 0].sort_values(ascending=False)
```

```python
cols = ['Communication_Preference',
        "Favorite_Product_Category ",
        "Special_Notes "
]

df[col] = df[col].bfill()
```

```python
df.isna().sum()[df.isna().sum() > 0].sort_values(ascending=False)
```

<div dir="rtl" style="text-align: right;">


 </div>

```python
cols = ['Avg_Session_Duration_Minutes', 
        'Shipping_Distance_KM', 
        'Total_Refund_Amount','Total_Clicks_On_Ads',
        'Loyalty_Points',
        'Avg_Discount_Used',
        'Complaints_Count']

for col in cols:
    df[col]= df[col].fillna(df[col].mean())

df.isna().sum()[df.isna().sum() > 0].sort_values(ascending=False)
```

<div dir="rtl" style="text-align: right;">

3.9 إزالة القيم الفارغة أو النصوص غير صالحة
 </div>

```python
df['Phone_Number'] = df['Phone_Number'].replace(['Na--', 'nan--', 'None', '<NA>', 'NaN'], pd.NA)

df['Phone_Number']
```

<div dir="rtl" style="text-align: right;">
3.10 ازالة كل الحروف والنقاط والشرطات والأقواس، مع الحفاظ على الأرقام فقط

 </div>

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(f'\D', '', regex=True)
df['Phone_Number']
```

<div dir="rtl" style="text-align: right;">
3.11 إزالة الامتدادات بعد x إذا وجدت

 </div>

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'x\d+$', '', regex=True)
df['Phone_Number']
```

<div dir="rtl" style="text-align: right;">

3.12 إزالة الفراغات الزائدة في البداية أو النهاية
 </div>

```python
df['Phone_Number'] = df['Phone_Number'].str.strip()
df['Phone_Number']
```

<div dir="rtl" style="text-align: right;">

3.13 التعامل مع القيم الفارغة <NA>
 </div>

```python
df['Phone_Number'] = df['Phone_Number'].fillna('Unknown')
df['Phone_Number']
```

<div dir="rtl" style="text-align: right;">
3.14 قص أو التحقق من طول الرقم

 </div>

```python
df['Phone_Number'] = df['Phone_Number'].where(df['Phone_Number'].str.len().between(10,15), 'Invalid')
df['Phone_Number']
```
