0. قراءة الملف

```python
df = pd.read_csv(r"D:\Pandas\Pandas_Course\world_population.csv")
```


<div dir="rtl" style="text-align: right;">
1. عرض البيانات

 </div>

```python
df
```


<div dir="rtl" style="text-align: right;">
2. تصفية الصفوف (Rows) بحسب شرط معين
 </div>

```python
df[df["Rank"] <=10]
```


<div dir="rtl" style="text-align: right;">
3. تصفية الصفوف (Rows) بحسب  تصفية بشرطين معاً
 </div>

```python
df_filter = df[(df['Continent'] == 'Asia') & (df['2022 Population'] > 50_000_000)]
df_filter
```


<div dir="rtl" style="text-align: right;">

4. التصفية باستخدام OR

 </div>

```python
df_or = df[(df['Continent'] == 'Asia') | (df['Continent'] == 'Africa')]
df_or
```


<div dir="rtl" style="text-align: right;">

5. التصفية بالنفي
 </div>

```python
df_not = df[~df['Continent'].isin(['Asia','Africa'])]
df_not
```


<div dir="rtl" style="text-align: right;">
6. تصفية الصفوف (Rows) بحسب قيم محددة 
 </div>

```python
specific_countries = ['Bangladesh','Brazil']

df[df['Country'].isin(specific_countries)]
```


<div dir="rtl" style="text-align: right;">
7. فلترة الأسطر بحسب البحث عن قيمة معينة داخل عمود معين
</div>

```python
df[df['Country'].str.contains('United')]
```


<div dir="rtl" style="text-align: right;">
تعديل الفهرسة</div>

```python
df2 = df.set_index('Country')
df2
```


<div dir="rtl" style="text-align: right;">
8. تصفية الأعمدة واختيار عمودين فقط لإظهارهما
</div>

```python
df2.filter(items = ['Continent', 'CCA3'], axis=1)
```


<div dir="rtl" style="text-align: right;">
9. تصفية الأسطر  واختيار قيمة معينة (دولة) فقط لإظهارها
</div>

```python
df2.filter(items=['Zimbabwe'], axis=0)
```


<div dir="rtl" style="text-align: right;">

10. تصفية الصفوف بالبحث عن قيمة محددة في عمود الفهرسة (الذي هو عمود الدولة)

</div>

```python
df2.filter(like ='United', axis=0)
```


<div dir="rtl" style="text-align: right;">

11. الوصول المباشر لبيانات وفق قيمة موجودة ضمن عمود الفهرسة (الذي هو عمود الدولة)
</div>

```python
df2.loc['United States']
```


<div dir="rtl" style="text-align: right;">

12. الوصول المباشر لبيانات وفق رقم الصف عمود
</div>

```python
df2.iloc[3]
```


<div dir="rtl" style="text-align: right;">

13. تصفية الصفوف وفق شرط معين و ترتيب تصاعدي للبيانات

</div>

```python
df_filtered= df[df['Rank'] < 10].sort_values(by=['Continent', 'Country'], ascending=[False,True])
df_filtered
```
